```python
import numpy as np
from threading import Thread,Lock,RLock
import time
from queue import Queue
```

```python
class Tortuga(Thread):
    def __init__(self):
        super().__init__()
    
    def suceso(self):
        global pocision_t
        global gano_tortuga
        global tiempo_llegada_t
        global tiempo_llegada_l
        num=np.random.randint(0,100)
        
        if(pocision_t>=70):
            tiempo_llegada_t=time.time()
            if(tiempo_llegada_l==tiempo_llegada_t):
                print("Empataron")
            else:
                print("Gano tortuga!")
            gano_tortuga=True
        elif(num <50):
            if((pocision_t+3)>=70):
                pocision_t=70
            else:
                pocision_t+=3
        elif(num<70):
            if(pocision_t<=7):
                pocision_t=1
            else:
                pocision_t-=6
        else:
            pocision_t+=1

    def run(self):
        global pocision_t
        global gano_tortuga
        global gano_liebre
        print(f"{pocision_t}-T")
        while not(gano_tortuga):
            if not(gano_liebre):
                self.suceso()
                print(f"{pocision_t}-T \n")
                time.sleep(1)
                
            else:
                break
```

```python
class Liebre(Thread):
    def __init__(self):
        super().__init__()
    
    def suceso(self):
        global pocision_l
        global gano_liebre
        global tiempo_llegada_l
        global tiempo_llegada_t
        num=np.random.randint(0,100)
        
        if(pocision_l>=70):
            gano_liebre=True
            tiempo_llegada_l=time.time()
            if(tiempo_llegada_l==tiempo_llegada_t):
                print("Empataron")
            else:
                print("Gano liebre!")
            
        elif(num <20):
            pass
        elif(num<40):
            if(pocision_l+9>=70):
                pocision_l=70
            else:
                pocision_l+=9
            
        elif(num<50):
            if(pocision_l<=13):
                pocision_l=1
            else:
                pocision_l-=12
                
        elif(num<80):
            pocision_l+=1
            
        else:
        
            if(pocision_l<=3):
                pocision_l=1
            else:
                pocision_l-=2

    def run(self):
        global pocision_l
        global gano_tortuga
        global gano_liebre
        
        print(f"{pocision_l}-L")
        while not(gano_liebre):
            if not(gano_tortuga):
                
                self.suceso()
                print(f"{pocision_l}-L \n")
                time.sleep(1)
            else:
                break
      
```

```python
if __name__=="__main__":
    pocision_t=1
    pocision_l=1
    gano_tortuga=False
    gano_liebre=False
    tiempo_llegada_t=0
    tiempo_llegada_l=0
    
    tortuga=Tortuga()
    liebre=Liebre()
    print("Empieza la carrera!!")
    tortuga.start()
    liebre.start()
    
    tortuga.join()
    liebre.join()
```
